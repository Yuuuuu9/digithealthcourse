### SQL Format

FROM [table_name]\
WHERE [conditions];\
GROUP BY [column_name]\
ORDER BY [column_name]\
LIMIT

### 2. Using multiple conditions

````
%%bigquery --project mysites-248817  #replace with your project id
SELECT * FROM `physionet-data.mimiciii_clinical.patients` WHERE gender='F' and SUBJECT_ID = 249
````

### 3. COUNT

````

SELECT COUNT(*) FROM `physionet-data.mimiciii_clinical.patients` WHERE gender='F'
````

### 4. JOINS

````
%%bigquery --project mysites-248817
SELECT SUBJECT_ID,HADM_ID, d.ICD9_CODE, SHORT_TITLE from `physionet-data.mimiciii_clinical.diagnoses_icd` as d LEFT JOIN `physionet-data.mimiciii_clinical.d_icd_diagnoses` as dd ON d.ICD9_CODE=dd.ICD9_CODE WHERE d.SUBJECT_ID = 12
````

### 5. GROUP BY

````
%%bigquery --project mysites-248817
SELECT SUBJECT_ID,COUNT(HADM_ID) as HADM_COUNT from `physionet-data.mimiciii_clinical.admissions` GROUP BY SUBJECT_ID HAVING COUNT(HADM_ID)>=2
````
